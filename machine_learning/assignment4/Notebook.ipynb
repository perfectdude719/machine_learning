## Experiment 1 : Clustering without PCA

```python
import numpy as np

def kmeans_hill_climbing(data, k, max_iter=100, restarts=10):
    best_labels = None
    best_centroids = None
    best_sse = float('inf')  # Initialize with a very high SSE value

    for restart in range(restarts):
        # Step 1: Randomly initialize centroids
        np.random.seed(42 + restart)  # Change seed for different random restarts
        random_indices = np.random.choice(data.shape[0], k, replace=False)
        centroids = data[random_indices]
        
        for i in range(max_iter):
            # Step 2: Assign points to nearest centroid
            distances = np.linalg.norm(data[:, None] - centroids, axis=2)
            labels = np.argmin(distances, axis=1)
            
            # Step 3: Recalculate centroids
            new_centroids = np.array([data[labels == j].mean(axis=0) for j in range(k)])
            
            # Check for convergence
            if np.all(centroids == new_centroids):
                break
            centroids = new_centroids
        
        # Compute SSE for the current clustering
        sse = np.sum((data - centroids[labels]) ** 2)
        
        # Update the best solution if current SSE is lower
        if sse < best_sse:
            best_sse = sse
            best_labels = labels
            best_centroids = centroids
    
    return best_labels, best_centroids, best_sse

# Example: Cluster into 3 clusters with 10 restarts
labels, centroids, sse = kmeans_hill_climbing(data_scaled.values, k=2, restarts=10)
print(f"Best SSE: {sse}")
```

 Tunning the parameter k

```python
# Grid search over k
def tune_k(data, k_values, max_iter=100):
    results = []
    for k in k_values:
        _, _, sse = kmeans_hill_climbing(data, k, max_iter)
        results.append((k, sse))
        print(f"k: {k}, SSE: {sse}")
    return results

# Define the range of k values to test
k_values = range(1, 11)  # Test k from 1 to 10

# Perform grid search
results = tune_k(data_scaled.values, k_values)

# Find the best k
best_k, best_sse = min(results, key=lambda x: x[1])
print(f"Best k: {best_k} with SSE: {best_sse}")

# Plot the results (optional)
import matplotlib.pyplot as plt
k_values, sse_values = zip(*results)
plt.plot(k_values, sse_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()
```

## Experiment 2 : Implementing PCA

```python
def pca(data, n_components):
    # Step 1: Compute the covariance matrix
    cov_matrix = np.cov(data.T)
    #print(cov_matrix)

    
    # Step 2: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # Step 3: Sort eigenvectors by descending eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    
    # Step 4: Select the top n_components
    top_eigenvectors = eigenvectors[:, :n_components]
    
    # Step 5: Project the data
    reduced_data = np.dot(data, top_eigenvectors)
    return reduced_data, eigenvalues, eigenvectors

# Example: Reduce to 2 components
reduced_data, eigenvalues, eigenvectors = pca(data_scaled.values, n_components=2)
#print(f"the eigenvalues are{eigenvalues}\n the eigen vectors are{eigenvectors}\n")
print(f"the first few samples of data are {reduced_data[:5]}")
```

```python
def against_true_label(data, true_labels=None, title="Cluster Visualization", centroids=None):
    # Check if data is 2D or 1D
    if data.shape[1] >= 2:
        # If data has 2 or more dimensions, plot true labels
        plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='coolwarm', label='True Labels', marker='x', s=100)  # Color by true labels
    else:
        # If data has only 1 dimension, plot true labels along a horizontal axis
        plt.scatter(data[:, 0], [0] * len(data), c=true_labels, cmap='coolwarm', label='True Labels', marker='x', s=100)  # Color by true labels

    # If centroids are provided, plot them as well
    if centroids is not None:
        if centroids.shape[1] >= 2:  # Ensure centroids have at least 2 dimensions
            plt.scatter(centroids[:, 0], centroids[:, 1], c='green', s=200, marker='X', label='Centroids')  # Centroids color
        elif centroids.shape[1] == 1:  # If centroids have only 1 dimension
            plt.scatter(centroids[:, 0], [0] * len(centroids), c='green', s=200, marker='X', label='Centroids')  # Centroids color

    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def new_visualize_clusters(data, labels, title, centroids=None):
    # Check the number of dimensions in the data
    if data.shape[1] >= 2:
        # If data has 2 or more dimensions, plot normally
        plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', label='Clusters')
    else:
        # If data has only 1 dimension, plot in a single line (for visualization)
        plt.scatter(data[:, 0], [0] * len(data), c=labels, cmap='viridis', label='Clusters')

    # If centroids are provided, plot them as well
    if centroids is not None:
        if centroids.shape[1] >= 2:  # Ensure centroids have at least 2 dimensions
            plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
        elif centroids.shape[1] == 1:  # If centroids have only 1 dimension
            plt.scatter(centroids[:, 0], [0] * len(centroids), c='red', s=200, marker='X', label='Centroids')

    plt.title(title)
    plt.colorbar(label='Cluster')
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from collections import defaultdict

def grid_search_pca_kmeans(data, k_range, max_components):
    sse_values = []
    best_k = None
    best_n_components = None
    sse_by_k = defaultdict(list)  # Dictionary to store SSE values grouped by k

    for k in k_range:
        for n_components in range(1, max_components + 1):
            # Perform PCA with n_components
            reduced_data, _, _ = pca(data, n_components)
            
            # Apply KMeans clustering to the reduced data and calculate SSE
            _, _, sse = kmeans_hill_climbing(reduced_data, k=k)  # Use k from the loop
            
            # Store SSE, k, and n_components
            sse_values.append((sse, k, n_components))
            sse_by_k[k].append((sse, n_components))
    
    # Sort each k's SSE list and take the top 2 combinations with the least SSE
    for k, sse_list in sse_by_k.items():
        sse_by_k[k] = sorted(sse_list, key=lambda x: x[0])[:2]

    # Print the top 2 combinations with the least SSE for each k
    print("\nTop 2 combinations with the least SSE for each k:")
    for k, top_combinations in sse_by_k.items():
        for sse, n_components in top_combinations:
            print(f"k={k}, n_components={n_components}, SSE={sse:.4f}")
            plot_reduced_data, _, _ = pca(data, n_components)
            labels, centroids, _ = kmeans_hill_climbing(plot_reduced_data, k=k, restarts=10)
            new_visualize_clusters(plot_reduced_data,labels,f"{k} clusters and {n_components} components",centroids=centroids)
            against_true_label(plot_reduced_data, true_labels=true_labels, title=f"{k} clusters and {n_components} components", centroids=centroids)
    
    # Find the combination with the minimal SSE
    min_sse, best_k, best_n_components = min(sse_values, key=lambda x: x[0])  # x[0] is the SSE value

    print(f"\nThe minimal SSE is {min_sse:.4f} with k={best_k} clusters and {best_n_components} PCA components.")
    
    # Prepare data for plotting
    sse_list, k_list, n_components_list = zip(*sse_values)
    
    # Create 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(k_list, n_components_list, sse_list, c=sse_list, cmap='viridis')

    ax.set_xlabel('Number of Clusters (k)')
    ax.set_ylabel('Number of PCA Components')
    ax.set_zlabel('SSE')

    plt.title('SSE vs. Number of Clusters and PCA Components')
    plt.colorbar(scatter, label="SSE")
    plt.show()
    
    return min_sse, best_k, best_n_components
```

```python
import numpy as np

def adjusted_kmeans(data, k, max_iter=100, tol=1e-4):
    # Step 1: Randomly initialize centroids
    np.random.seed(42)
    random_indices = np.random.choice(data.shape[0], k, replace=False)
    centroids = data[random_indices]
    
    for i in range(max_iter):
        # Step 2: Assign points to nearest centroid
        distances = np.linalg.norm(data[:, None] - centroids, axis=2)  # Corrected distance calculation
        labels = np.argmin(distances, axis=1)
        
        # Step 3: Recalculate centroids
        new_centroids = np.array([data[labels == j].mean(axis=0) if len(data[labels == j]) > 0 else data[np.random.choice(data.shape[0])] for j in range(k)])
        
        # Check for convergence (using tolerance)
        if np.all(np.abs(centroids - new_centroids) < tol):
            break
        centroids = new_centroids
    
    # Compute SSE (Sum of Squared Errors)
    sse = np.sum((data - centroids[labels]) ** 2)
    
    return labels, centroids, sse
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics import adjusted_rand_score, accuracy_score
import matplotlib.pyplot as plt

# Function to map predicted clusters to true labels
def map_clusters_to_labels(pred_labels, true_labels):
    mapped_labels = np.zeros_like(pred_labels)
    for i in np.unique(pred_labels):
        # Get indices of samples in the current cluster
        cluster_indices = np.where(pred_labels == i)[0]
        # Find the majority label in this cluster
        if len(cluster_indices) > 0:
            majority_label = np.argmax(np.bincount(true_labels[cluster_indices]))
            mapped_labels[cluster_indices] = majority_label
    return mapped_labels

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score, accuracy_score



def grid_search_pca_kmeans_ari(X, true_labels, n_components_list, k_list):
    results = []  # Store results for each combination of n_components and k
    
    for n_components in n_components_list:
        # Perform PCA using your custom function
        X_reduced, _, _ = pca(X, n_components)  # Using custom_pca instead of pca
        
        for k in k_list:
            # Perform K-means clustering (assuming adjusted_kmeans is defined)
            predicted_labels, _, _ = adjusted_kmeans(X_reduced, k=k)
            
            # Map clusters to true labels
            mapped_labels = map_clusters_to_labels(predicted_labels, true_labels)
            
            # Compute ARI and accuracy
            ari = adjusted_rand_score(true_labels, mapped_labels)
            accuracy = accuracy_score(true_labels, mapped_labels)
            
            # Append results
            results.append((n_components, k, ari, accuracy))
            print(f"PCA Components: {n_components}, K: {k}, ARI: {ari:.4f}, Accuracy: {accuracy:.4f}")
    
    # Find the best configuration based on accuracy
    best_n_components, best_k, best_ari, best_accuracy = max(results, key=lambda x: x[3])
    
    # Perform PCA for the best configuration
    X_reduced_best, _, _ = pca(X, best_n_components)  # Using custom_pca instead of pca
    
    # Plot the most accurate PCA (using the best number of components)
    plt.figure(figsize=(8, 6))
    
    if best_n_components == 1:
        # If we have only one component, plot in 1D
        plt.scatter(X_reduced_best, np.zeros_like(X_reduced_best), c=true_labels, cmap='viridis', edgecolor='k', s=50)
        plt.title(f"Best PCA: {best_n_components} Component, K={best_k} Clusters (Accuracy: {best_accuracy:.4f})")
        plt.xlabel('PCA Component 1')
        plt.ylabel('Zeroed for visualization')
    else:
        # If we have two components, plot in 2D
        plt.scatter(X_reduced_best[:, 0], X_reduced_best[:, 1], c=true_labels, cmap='viridis', edgecolor='k', s=50)
        plt.title(f"Best PCA: {best_n_components} Components, K={best_k} Clusters (Accuracy: {best_accuracy:.4f})")
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
    
    plt.colorbar(label='True Labels')
    plt.show()
    
    return best_n_components, best_k, best_ari, best_accuracy, results

# Example usage
n_components_list = range(1, 11)  # Test PCA components from 1 to 10
k_list = range(2, 6)  # Test K values from 2 to 5

best_n_components, best_k, best_ari, best_accuracy, results = grid_search_pca_kmeans_ari(data_scaled.values, true_labels, n_components_list, k_list)

# Extract results for plotting
results_array = np.array(results)
components = results_array[:, 0].astype(int)
ks = results_array[:, 1].astype(int)
accuracies = results_array[:, 3].astype(float)

# Plot heatmap of accuracy for PCA components vs. number of clusters (k)
plt.figure(figsize=(10, 8))
for k in k_list:
    plt.plot(
        components[ks == k], 
        accuracies[ks == k], 
        marker='o', 
        linestyle='-', 
        label=f'K={k}'
    )

plt.title('PCA Components vs Accuracy for Different K')
plt.xlabel('Number of PCA Components')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

print("\nSummary of Results:")
for n_components, k, ari, accuracy in results:
    print(f"PCA Components: {n_components}, K: {k}, ARI: {ari:.4f}, Accuracy: {accuracy:.4f}")

print(f"\nBest Configuration:")
print(f"Best number of PCA components: {best_n_components}")
print(f"Best number of clusters (K): {best_k}")
print(f"Best ARI: {best_ari:.4f}")
print(f"Best Accuracy: {best_accuracy:.4f}")
```

## extra ---> data reconstruction

```python


def pca_with_reconstruction(data, n_components):
    # Step 1: Compute the covariance matrix
    cov_matrix = np.cov(data.T)
    
    # Step 2: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # Step 3: Sort eigenvectors by descending eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    
    # Step 4: Select the top n_components
    top_eigenvectors = eigenvectors[:, :n_components]
    
    # Step 5: Project the data
    reduced_data = np.dot(data, top_eigenvectors)
    
    # Step 6: Reconstruct the original data
    reconstructed_data = np.dot(reduced_data, top_eigenvectors.T)
    
    return reduced_data, reconstructed_data, eigenvalues, eigenvectors


from sklearn.metrics import mean_squared_error

# Apply PCA and reconstruct the data  and apply some useful metrics
reduced_data, reconstructed_data, eigenvalues, eigenvectors = pca_with_reconstruction(data_scaled.values, n_components=2)

mse = mean_squared_error(data_scaled.values, reconstructed_data)
print("Mean Squared Error (MSE):\n", mse)

relative_error = np.linalg.norm(data_scaled.values - reconstructed_data) / np.linalg.norm(data_scaled.values)
print("Relative Reconstruction Error:", relative_error)


# Print the first 5 rows of the reduced and reconstructed data
print("\nReduced Data (First 2 Rows):")
print(reduced_data[:2])

print("\nReconstructed Data (First 2 Rows):")
print(reconstructed_data[:2])

print("\n Original Data (First 2 Rows):")
print(data_scaled.values[:2])
```
